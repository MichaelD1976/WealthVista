#### Access Oil Price


```python
import yfinance as yf
import pandas as pd
from pathlib import Path
```

```python
# Get the root of your project 
project_root = Path.cwd().parents[1]

# Construct the full path to data/raw
data_dir_to_save = project_root / 'data' / 'raw'
data_dir_to_save.mkdir(parents=True, exist_ok=True)  # creates dirs if needed

SAVE_FILE_PATH = data_dir_to_save / 'oil_prices_raw.csv'
```

```python
# Define the ticker symbol for gold futures
oil = yf.download("CL=F", start="2010-01-01", interval="1mo")
```

```python
# Download monthly gold prices since 2010-01-01
oil.tail(50)
```

```python
# If multi-index columns, extract Close price
if isinstance(oil.columns, pd.MultiIndex):
    oil = oil['Close']['CL=F']
else:
    oil = oil['Close']
```

```python
# Convert to DataFrame, reset index
oil_df = oil.to_frame(name='Oil_Price_USD').reset_index()
```

```python
oil_df.tail(50)
```

```python
# Round close price
oil_df['Oil_Price_USD'] = oil_df['Oil_Price_USD'].round(1)
```

```python
oil_df.to_csv(SAVE_FILE_PATH, index=False)
```
